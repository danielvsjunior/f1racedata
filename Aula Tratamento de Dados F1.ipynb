# Importando Biblioteca Pandas

```python
import pandas as pd
```

# Extraindo os dados desejados de cada arquivo

## Lista de Circuitos

```python
# Trazendo os Dados para o Dataframe

circuito = pd.read_csv('dados/circuits.csv',encoding = "ISO-8859-1")
circuito.head()
```

```python
## Deste Dataframe eu só vou utilizar 3 colunas: Referência, Nome do Circuito e o País

circuito = circuito[['circuitId','name','country']]
circuito.head()
```

### Vamos relacionar as colunas 'name' e 'country' com a coluna 'circuitId' em dois dicionários diferentes

```python
## Relacionando a coluna 'name' com a coluna 'circuitId' em um dicionário

circ_name = dict([(i,a) for i,a in zip(circuito['circuitId'],circuito['name'])])
circ_name
```

```python
## Relacionando a coluna 'country' com a coluna 'circuitId' em um dicionário

circ_pais = dict([(i,a) for i,a in zip(circuito['circuitId'],circuito['country'])])
circ_pais
```

## Lista de Pilotos

```python
pilotos = pd.read_csv('dados/drivers.csv',encoding = "ISO-8859-1")
pilotos.head()
```

```python
## No caso dos pilotos temos os nomes separados em duas colunas, vamos contatená-las em uma nova coluna para termos o nome completo de cada piloto

pilotos['piloto'] = pilotos['forename'] + ' ' + pilotos['surname']
pilotos.head()
```

```python
## Para o dicionário de Pilotos vamos utilizar apenas as colunas 'driverId' e a nova coluna 'piloto' que criamos

pilotos = pilotos[['driverId','piloto']]
pilotos.head()
```

### Vamos relacionar a coluna 'piloto' com a coluna 'driverId' em um dicionário

```python
## Vamos utilizar uma função diferente da anterior, pois neste caso temos apenas duas colunas

driver = pilotos.set_index('driverId').T.to_dict('list')
driver
```

## Lista de Corridas

```python
corridas = pd.read_csv('dados/races.csv')
corridas.head()
```

```python
## Para este dicionário vamos utilizar apenas as colunas 'raceId', 'year' e 'circuitId'

corridas = corridas[['raceId','year','circuitId']]
corridas.head()
```

### Vamos relacionar as colunas 'year' e 'circuitId' com a coluna 'raceId' em dois dicionários diferentes

```python
## Relacionando a coluna 'year' com a coluna 'raceId' em um dicionário

race_year = dict([(i,a) for i,a in zip(corridas['raceId'],corridas['year'])])
race_year
```

```python
## Relacionando a coluna 'circuitId' com a coluna 'raceId' em um dicionário

race_circ = dict([(i,a) for i,a in zip(corridas['raceId'],corridas['circuitId'])])
race_circ
```

## Tabela de Classificação

```python
quali = pd.read_csv('dados/qualifying.csv',encoding = "ISO-8859-1")
quali.head()
```

```python
## Nesta Dataframe não iremos utilizar as colunas 'constructorId' e 'number' e manteremos as demais

quali = quali.drop(columns=['constructorId','number'])
quali.head()
```

# Inserindo os Dados Extraídos na Tabela de Classificação

#### Note que neste Dataframe temos as referências 'raceId' e 'driverId', que iremos preencher com os dados dos respectivos dicionários criados

```python
# Primeiro vamos substituir os dados da coluna 'driverId' pelos dados do dicionário driver que criamos anteriormente

quali['driverId'] = quali['driverId'].replace(driver)
quali.head()
```

```python
# Note que no dataframe há apenas uma coluna 'raceId', porém temos dois dicionários relacionados à essa referência.
# Para resolvermos isso temos que duplicar a coluna 'raceId', uma irá receber os dados do ano da corrida e outra irá receber
# o nome do circuito.

quali['circuito'] = quali['raceId']
quali.head()
```

```python
# Agora podemos fazer a substituição

quali['raceId'] = quali['raceId'].replace(race_year)
quali['circuito'] = quali['circuito'].replace(race_circ)
quali.head()
```

```python
# Precisamos duplicar também a coluna 'circuito' pois também temos  dois dicionários para cada circuito

quali['pais'] = quali['circuito']
quali.head()
```

```python
# Agora podemos fazer a substituição da mesma maneira

quali['circuito'] = quali['circuito'].replace(circ_name)
quali['pais'] = quali['pais'].replace(circ_pais)

quali.head()
```

```python
# Vamos renomear as colunas
quali = quali.rename(columns={
    "qualifyId":"Quali ID",
    "raceId":"Ano",
    "driverId":"Piloto",
    "position":"Pos",
    "circuito":"Circuito",
    "pais":"País",
    "q1":"Q1",
    "q2":"Q2",
    "q3":"Q3"})
quali.head()
```

```python
# Agora vamos reorganizar as colunas para ficar mais fácil a visualização

quali = quali.reindex(columns = ['Quali ID','Circuito','País','Ano','Piloto','Pos','Q1','Q2','Q3'])
quali.head()
```

```python
# Por fim, vamos salvar este dataframe como um arquivo .csv para podermos manipulá-lo novamente no futuro

quali.to_csv('dados/tabela_de_classificacao.csv', sep = ';', index = False)
```
